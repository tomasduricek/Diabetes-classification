```python
# Detect outliers
def outlier(data):
    data=np.sort(data)
    Q1, Q3 = np.percentile(data, [25, 75])
    IQR = Q3 - Q1
    lower = Q1 - IQR * 1.5
    upper = Q3 + IQR * 1.5
    return lower, upper
```

```python
lower, upper  = outlier(df['SerumInsulin'])
print(lower, upper)
```

```python
# More than 650 outliers. Did not deleted this feature, because it can be significant in classification
df[df['SerumInsulin'] > upper]
```

```python
### Split data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Logistic Regression

```python
lg = LogisticRegression()

params_lg = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'multi_class': ['auto', 'ovr', 'multinomial']
}
grid_lg = GridSearchCV(estimator = lg, param_grid=params_lg, cv=3, verbose=1)
model_lg = grid_lg.fit(X_train, y_train)
y_pred_lg = model_lg.predict(X_test)
print(f'Precision: {precision_score(y_test, y_pred_lg)} \nAccuracy: {accuracy_score(y_test, y_pred_lg)} \nRecall: {recall_score(y_test, y_pred_lg)}\nF1 Score: {f1_score(y_test, y_pred_lg)}')
plot_confusion_matrix(model_lg, X_test, y_test)
```

### Random Forest

```python
rfc = RandomForestClassifier()

params_rfc = {
    'n_estimators': list(range(10,100,5)),
    'max_features': ['auto', 'sqrt'],
    'max_depth': range(2, 10)
}
grid_rfc = GridSearchCV(estimator = rfc, param_grid=params_rfc, cv=3, verbose=1)
model_rfc = grid_rfc.fit(X_train, y_train)
y_pred_rfc = model_rfc.predict(X_test)
print(f'Precision: {precision_score(y_test, y_pred_rfc)} \nAccuracy: {accuracy_score(y_test, y_pred_rfc)} \nRecall: {recall_score(y_test, y_pred_rfc)}\nF1 Score: {f1_score(y_test, y_pred_rfc)}')
plot_confusion_matrix(model_rfc, X_test, y_test)
```

### Support Vector Machine

```python
svc = SVC()

params_svc = {'C': [0.1, 1, 10, 100], 
              'gamma': [1, 0.1, 0.01],
              'kernel': ['rbf']} 

grid_svc = GridSearchCV(estimator = svc, param_grid=params_svc, cv=3, verbose=1)
model_svc = grid_svc.fit(X_train, y_train)
y_pred_svc = model_svc.predict(X_test)
print(f'Precision: {precision_score(y_test, y_pred_svc)} \nAccuracy: {accuracy_score(y_test, y_pred_svc)} \nRecall: {recall_score(y_test, y_pred_svc)}\nF1 Score: {f1_score(y_test, y_pred_svc)}')
plot_confusion_matrix(model_svc, X_test, y_test)
```

### KNN

```python
# For KNN it is needed to standardize all columns
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier()

params_knn = { 'n_neighbors' : np.arange(3, 50, 2),
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}

grid_knn = GridSearchCV(estimator = knn, param_grid=params_knn, cv=3, verbose=1)
model_knn = grid_knn.fit(X_train_scaled, y_train)
y_pred_knn = model_knn.predict(X_test_scaled)
print(f'Precision: {precision_score(y_test, y_pred_knn)} \nAccuracy: {accuracy_score(y_test, y_pred_knn)} \nRecall: {recall_score(y_test, y_pred_knn)}\nF1 Score: {f1_score(y_test, y_pred_knn)}')
plot_confusion_matrix(model_knn, X_test_scaled, y_test)
```

### AdaBoost

```python
ada = AdaBoostClassifier()

params_ada = {'n_estimators': (100, 500, 50)} # (np.arange (400, 500, 5), 465 was the best

grid_ada = GridSearchCV(estimator = ada, param_grid=params_ada, cv=3, verbose=1)
model_ada = grid_ada.fit(X_train, y_train)
y_pred_ada = model_ada.predict(X_test)
print(f'Precision: {precision_score(y_test, y_pred_ada)} \nAccuracy: {accuracy_score(y_test, y_pred_ada)} \nRecall: {recall_score(y_test, y_pred_ada)}\nF1 Score:{f1_score(y_test, y_pred_ada)}')
fig, ax = plt.subplots(figsize=(6.4, 4.8))
plot_confusion_matrix(model_ada, X_test, y_test, ax=ax)
plt.show()
```

```python
f1_data = {'AdaBoost':f1_score(y_test, y_pred_ada), 'KNN':f1_score(y_test, y_pred_knn), 'Support Vector Classifier':f1_score(y_test, y_pred_svc), 'Random Forest Classifier':f1_score(y_test, y_pred_rfc),'Logistic Regression': f1_score(y_test, y_pred_lg)}
f1_scores = pd.Series(f1_data)
```

```python
plt.figure(figsize=(7, 3))
graph = sns.barplot(f1_scores, y=f1_scores.index, palette='rocket', )
plt.title('F1 Scores')
plt.show()
```
